```python
import pandas as pd

df = pd.read_csv('employee_salary_dataset.csv')
df.head()

```

```python
import numpy as np
```

```python
print("First 5 Rows")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

# 2. Handle Missing Values


```python
df.fillna(df.select_dtypes(include=np.number).mean(), inplace=True)
```

# 3. Remove Duplicate Records


```python
df.drop_duplicates(inplace=True)
```

Create Bonus Column


```python
df["Bonus"] = df["Monthly_Salary"] * 0.10
print("Dataset after Bonus Column")
print(df.head())
```

# 4. Create new  Column: Total Income


```python
df["Total_Income"] = df["Monthly_Salary"] + df["Bonus"]

print("\nDataset after Bonus and Total Income")
print(df.head())
```

#5. Find employees earning above the average salary.

```python
avg_salary = df["Monthly_Salary"].mean()

above_avg = df[df["Monthly_Salary"] > avg_salary]

print("\nAverage Salary =", avg_salary)

print("\nEmployees Earning Above Average Salary")
print(above_avg[["EmployeeID","Name","Department","Monthly_Salary"]])
```

#6. Find department-wise average salary.

```python
dept_avg = df.groupby("Department")["Monthly_Salary"].mean()

print("\nDepartment-wise Average Salary")
print(dept_avg)
```

#7. Apply NumPy functions (max, min, mean, std) on salaries.

```python
salary_array = df["Monthly_Salary"].to_numpy()

print("\nSalary Statistics")
print("Maximum Salary :", np.max(salary_array))
print("Minimum Salary :", np.min(salary_array))
print("Mean Salary :", np.mean(salary_array))
print("Standard Deviation :", np.std(salary_array))
```

#8.Find the highest-paid employee.

```python
highest_paid = df.loc[df["Monthly_Salary"].idxmax()]

print("\nHighest Paid Employee")
print(highest_paid)
```

#9.Generate a summary report.

```python
print("\n========== SUMMARY REPORT ==========")

print("Total Employees :", len(df))
print("Departments :", df["Department"].nunique())
print("Cities :", df["City"].nunique())

print("Average Salary :", round(avg_salary,2))
print("Highest Salary :", np.max(salary_array))
print("Lowest Salary :", np.min(salary_array))

print("Highest Paid Employee :", highest_paid["Name"])
print("Department :", highest_paid["Department"])

print("Average Experience :", round(df["Experience_Years"].mean(),2))
```
