A **stack** is a data structure that can hold many elements, and the last element added is the first one to be removed.

Basic operations of a stack:

1.Push: Adds a new element on the stack.

2.Pop: Removes and returns the top element from the stack.

3.Peek: Returns the top (last) element on the stack.

4.isEmpty: Checks if the stack is empty.

5.Size: Finds the number of elements in the stack.

Stacks can be implemented by using arrays or linked lists.

Stacks can be used to implement undo mechanisms, to revert to previous states, to create algorithms for depth-first search in graphs, or for backtracking.

**Applications**
1. Undo functionality
2. Browser history
3. Function call management
4. Expression evaluation


stack operations.png

```python
# Stack class implementation
class Stack:
    def __init__(self):
        # Create an empty list to store stack elements
        self.items = []

    # Add an element to the top of the stack
    def push(self, item):
        self.items.append(item)

    # Remove and return the top element
    def pop(self):
        if self.is_empty():
            print("Stack is empty")
            return None

        return self.items.pop()

    # Return the top element without removing it
    def peek(self):
        if self.is_empty():
            print("Stack is empty")
            return None

        return self.items[-1]

    # Check whether the stack is empty
    def is_empty(self):
        return len(self.items) == 0

    # Return the number of elements in the stack
    def size(self):
        return len(self.items)

    # Display the stack
    def display(self):
        print(self.items)


# Create a stack object
stack = Stack()

# Push elements onto the stack
stack.push(10)
stack.push(20)
stack.push(30)

# Display stack
stack.display()     # [10, 20, 30]

# View top element
print("Top:", stack.peek())   # 30

# Remove top element
print("Popped:", stack.pop()) # 30

# Display updated stack
stack.display()     # [10, 20]
```

**Queue (FIFO)**

FIFO = First In, First Out

The first element added is the first one removed.

**Applications**
1. CPU scheduling
2. Print queues
3. Task processing
4. Breadth-First Search (BFS)

```python
stack=[]
stack.append(10)
stack.append(20)
print(stack)
stack.pop()
print("after removing",stack)
```

Queue operations.png

```python
# Queue class implementation
class Queue:
    def __init__(self):
        # Create an empty list to store queue elements
        self.items = []

    # Add an element to the rear (end) of the queue
    def enqueue(self, item):
        self.items.append(item)

    # Remove and return the front element
    def dequeue(self):
        # Check if queue is empty
        if self.is_empty():
            print("Queue is empty")
            return None

        # Remove the first element (front)
        return self.items.pop(0)

    # Return the front element without removing it
    def front(self):
        if self.is_empty():
            print("Queue is empty")
            return None

        return self.items[0]

    # Check whether the queue is empty
    def is_empty(self):
        return len(self.items) == 0

    # Return the number of elements in the queue
    def size(self):
        return len(self.items)

    # Display queue contents
    def display(self):
        print(self.items)


# Create a queue object
queue = Queue()

# Add elements to the queue
queue.enqueue(10)
queue.enqueue(20)
queue.enqueue(30)

queue.display()      # [10, 20, 30]

# View front element
print("Front:", queue.front())   # 10

# Remove front element
print("Dequeued:", queue.dequeue())  # 10

queue.display()      # [20, 30]
```

**Linked** List

A linked list consists of nodes, where each node contains:

1. Data
2. Reference (pointer) to the next node

```python
# Represents a single node in the linked list
class Node:
    def __init__(self, data):
        # Store the actual value/data
        self.data = data

        # Reference to the next node
        # Initially there is no next node
        self.next = None


# Represents the entire linked list
class LinkedList:
    def __init__(self):
        # The first node of the list
        # Initially the list is empty
        self.head = None

    # Add a new node at the end of the list
    def append(self, data):

        # Create a new node containing the given data
        new_node = Node(data)

        # If the list is empty,
        # make the new node the head node
        if not self.head:
            self.head = new_node
            return

        # Start from the head node
        current = self.head

        # Move through the list until reaching
        # the last node (whose next is None)
        while current.next:
            current = current.next

        # Link the last node to the new node
        current.next = new_node

    # Display all elements in the linked list
    def display(self):

        # Start traversal from the head
        current = self.head

        # Continue until there are no more nodes
        while current:

            # Print current node's data
            print(current.data, end=" -> ")

            # Move to the next node
            current = current.next

        # Indicates the end of the linked list
        print("None")
ll = LinkedList()

ll.append(10)
ll.append(20)
ll.append(30)

ll.display()
```

**Tree**

A Tree is a hierarchical data structure made up of nodes connected by edges.

Unlike a linked list (which is linear), a tree branches out.

**Why Trees Are Important**

Trees are used everywhere:

1. File systems (Documents → Projects → Files)

2. Database indexes (B-Trees)

3. HTML DOM structure

4. Organization charts

5. Decision trees in Machine Learning

6. Expression trees in compilers

```python
# Represents a single node in the tree
class Node:
    def __init__(self, data):
        # Value stored in the node
        self.data = data

        # Reference to left child
        self.left = None

        # Reference to right child
        self.right = None


# Create nodes
root = Node(10)

root.left = Node(5)
root.right = Node(20)

root.left.left = Node(3)
root.left.right = Node(8)

root.right.right = Node(30)
```

```python
'''1. Inorder Traversal
Left → Root → Right'''
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)

inorder(root)
```

```python
'''Preorder Traversal
Root → Left → Right'''
def preorder(node):
    if node:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)

preorder(root)
```

```python
'''Postorder Traversal
Left → Right → Root'''
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=" ")

postorder(root)
```
