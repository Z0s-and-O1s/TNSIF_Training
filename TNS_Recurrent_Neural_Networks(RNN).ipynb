#Recurrent Neural Networks (RNN)

Recurrent Neural Network (RNN) is a type of neural network specially designed to work with sequential data.

Unlike CNNs that process images, RNNs process data one step at a time while remembering previous information.

Definition

RNN (Recurrent Neural Network) is a neural network that remembers previous inputs using a hidden state, making it suitable for sequence and time-series data.

#Why do we need RNN?

Traditional Neural Networks treat every input independently.

Example:

Sentence:

"I am learning Deep Learning."

A normal neural network processes each word separately.

It cannot understand that

"Learning" comes after "Deep."

RNN remembers previous words.

Therefore it understands context.

#Real Life Analogy

Imagine reading a story.

You don't forget the previous page when reading the next page.

You remember what happened earlier.

Similarly,

RNN remembers previous inputs while processing the next input.

Where is RNN used?

RNN is mainly used wherever order matters.

Examples
- Text Prediction
- Machine Translation
- Speech Recognition
- Chatbots
- Time Series Forecasting
- Stock Price Prediction
- Weather Forecasting
- Sentiment Analysis





#Difference Between ANN, CNN and RNN
| ANN                | CNN                 | RNN                       |
| ------------------ | ------------------- | ------------------------- |
| Tabular Data       | Images              | Sequential Data           |
| No Memory          | Detects Features    | Remembers Previous Inputs |
| Independent Inputs | Spatial Information | Temporal Information      |

#Advantages of RNN

✅ Works well on sequential data

✅ Remembers previous information

✅ Suitable for NLP tasks

✅ Good for speech recognition

✅ Good for time-series prediction

#Disadvantages of RNN

❌ Slow training

❌ Difficult to train on long sequences

❌ Vanishing Gradient Problem

❌ Forgets old information over long sequences


#RNN Workflow

Sequential Data

      │
      ▼
Input Sequence (x₁, x₂, x₃, ...)

      │
      ▼
Initialize Hidden State (h₀ = 0)

      │
      ▼
Process First Input (x₁)

      │
      ▼
Generate Hidden State (h₁) + Output (y₁)

      │
      ▼
Process Next Input (x₂) + Previous Hidden State (h₁)

      │
      ▼
Generate New Hidden State (h₂) + Output (y₂)

      │
      ▼
Repeat for All Inputs

      │
      ▼
Final Output / Prediction

- Input Sequence → Give the sequence (text, speech, or time-series data).
- Initialize Memory → Start with an empty hidden state (h₀ = 0).
- Process Current Input → Read one input at a time.
- Update Memory → Combine the current input with the previous hidden state to - create a new hidden state.
- Generate Output → Produce an output for the current step.
- Repeat → Continue until the entire sequence is processed.
- Final Prediction → Use the final output (or all outputs) to make the prediction.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Sample sequence data
X = np.array([
    [[1], [2], [3]],
    [[2], [3], [4]],
    [[3], [4], [5]],
    [[4], [5], [6]]
])

# Target values
y = np.array([4, 5, 6, 7])

# Build the RNN model
model = Sequential([
    SimpleRNN(10, activation='tanh', input_shape=(3, 1)),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X, y, epochs=200, verbose=0)

# Test prediction
test = np.array([[[5], [6], [7]]])
prediction = model.predict(test)

print("Predicted Next Number:", prediction[0][0])
#print("Predicted Next Number:", prediction[0][1])
#print("Predicted Next Number:", prediction[0][2])
```
