##What is a Neural Network?
Definition

A Neural Network is a computer model inspired by the human brain. It consists of many interconnected artificial neurons that work together to learn patterns from data and make predictions.

Simple Definition

A Neural Network is a collection of artificial neurons that learn from data to solve problems like image recognition, speech recognition, and language translation.

Formula

A neural network is made up of many neurons. Each neuron performs the following calculation:

z=(x1​w1​+x2​w2​+x3​w3​+⋯+xn​wn​)+b​

Then an activation function is applied:

Output=f(z)

Where:

x = Input values (features)

w = Weights (importance of each input)

b = Bias (extra value to improve learning)

f = Activation Function

Output = Prediction

##Why do we need an Activation Function?
Definition

An Activation Function is a mathematical function that decides whether a neuron should be activated (produce an output) or not.

It introduces non-linearity into the neural network, allowing it to learn complex patterns.

Why is it needed?

Suppose a neuron calculates

z=x1w1+x2w2+b

If we directly use this value,

Input → Output

the entire neural network behaves like a simple linear equation.

It cannot learn:

Activation functions solve this problem by adding non-linearity.

#1. Sigmoid Activation Function
Definition

The Sigmoid function converts any input value into a number between 0 and 1.

It is mainly used when the output has only two possible classes.

Formula:

σ(x)=1/(1+e^−x)

- Output Range ->
0 to 1

- Interpretation ->If the output is 0.95

- It means-> 95% confidence

#​When to Use?

Use Sigmoid when solving Binary Classification problems.

Examples

Spam / Not Spam
Pass / Fail
Disease / No Disease
Fraud / Not Fraud
Cat / Not Cat


- Example

Email Spam Detection

Input->Email

Output->0.97

Meaning->97% Spam

If output > 0.5

Spam

Else

Not Spam


Study Hours

     │
     ▼
Multiply by Weight

     │
     ▼
Add Bias

     │
     ▼
Raw Score (z)

     │
     ▼
Sigmoid Function

     │
     ▼
Probability (0 to 1)

     │
     ▼
Final Prediction
(Pass or Fail)

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))  #where e = 2.718

study_hours = 5

weight = 2
bias = -8

z = (weight * study_hours) + bias

prediction = sigmoid(z)

print("Z Value :", z)
print("Sigmoid Output :", prediction)
```

#2. ReLU (Rectified Linear Unit)
Definition

ReLU returns the input if it is positive; otherwise, it returns 0.

It is the most commonly used activation function in Deep Learning.

Formula
f(x)=max(0,x)

- Output Range ->
0 to ∞

- Rule
1. Positive Number → Keep it

2. Negative Number → Make it Zero
Example

Input

-3

Output

0

Input

8

Output

8

##When to Use?

Use ReLU in Hidden Layers.

It works well for

CNN
ANN
Image Classification
Object Detection


- Real-Life Example

Imagine a motion sensor light.

Motion Detected

↓

Light ON

No Motion

↓

Light OFF

Negative values are ignored.

Positive values are allowed.

```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

weight = 2
bias = -8

hours = [1,2,3,4,5,6,7,8]

print("Hours\tZ\tReLU")

for h in hours:
    z = weight*h + bias
    output = relu(z)

    print(f"{h}\t{z}\t{output}")
```

#3. Tanh (Hyperbolic Tangent)
Definition

Tanh converts the input into values between -1 and 1.

Unlike Sigmoid, it can produce negative outputs.

Formula :    
tanh(x)=(e^x−e^−x​)/(e^x+e^−x)

Output Range ->
-1 to 1

##When to Use?

Use Tanh

Hidden Layers,
Recurrent Neural Networks (RNNs),
Time-series and sequence models

- Real-Life Example

Sentiment Analysis

Very Positive → +1

Neutral → 0

Very Negative → -1


```python
import numpy as np

def tanh(x):
    return np.tanh(x)

z_values = [-6, -2, 0, 2, 6]

outputs = tanh(z_values)

print("Neuron Outputs :", z_values)
print("After Tanh     :", outputs)
```

#4. Leaky ReLU
Definition

Leaky ReLU is an improved version of ReLU.

Instead of converting all negative values to 0, it keeps a very small negative value.

Formula :    
  f(x)={ x, x>0

  0.01x , x≤0

  - Output Range->(-∞,∞)

#Why was it introduced?

ReLU has a problem called the Dying ReLU Problem.

If a neuron keeps receiving negative inputs, its output becomes 0 repeatedly, and it may stop learning.

Leaky ReLU allows a small negative output (e.g., 1% of the input), helping the neuron continue learning.

#When to Use?

Use Leaky ReLU when

ReLU causes dead neurons,
Training very deep neural networks,
Building CNNs


#Real-Life Example

Imagine a water tap.

ReLU

Negative Input

↓

Water Completely OFF

Leaky ReLU

Negative Input

↓

Very Small Water Flow

The flow never becomes completely zero.


```python
import numpy as np

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

scores = [-10, -5, -2, 0, 2, 5, 10]

print("Score\tLeaky ReLU Output")

for score in scores:
    print(score, "\t", leaky_relu(score))
```

#5. Softmax
Definition

Softmax converts multiple output values into probabilities whose total is always 1.

It is used for Multi-Class Classification.

Formula :    
 P(xi) = e^xi / ∑n j=1 e^xj


-  Output Range->0 to 1

Total Probability = 1
#When to Use?

Use Softmax when there are more than two classes.

Examples

Cat
Dog
Horse

or

Apple
Mango
Orange
Banana

#Real-Life Example

Suppose an AI model identifies fruits.

Output

Apple

Banana

Orange

Softmax gives

Apple  → 0.82

Banana → 0.12

Orange → 0.06

The model predicts Apple because it has the highest probability.

```python
import numpy as np

# Raw scores (logits)
scores = np.array([4.5, 2.0, 0.5])

# Softmax function
def softmax(x):
    exp_values = np.exp(x)
    return exp_values / np.sum(exp_values)

probabilities = softmax(scores)

print("Softmax Probabilities:")
print(probabilities)

classes = ["Dog", "Cat", "Bird"]

predicted = classes[np.argmax(probabilities)]

print("\nPredicted Class :", predicted)
```
