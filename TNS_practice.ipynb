```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('netflix_titles.csv')
# print(data.head())
data.tail()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
# Which type of content is more available on Netflix?
type_counts = data['type'].value_counts()

# print(type_counts)

plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%')
plt.title("Content in Netflix")
plt.show()
```

```python
# Which countries contribute the most content to Netflix?
country_counts = data['country'].value_counts()

# print(country_counts)

df = country_counts.head(30).reset_index()
df.columns = ['country', 'count']

# plt.pie(country_counts[:10], labels=country_counts.index[:10], autopct='%1.1f%%')
# plt.title("Contribution in Netflix")
# plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='country', y='count', data=df, palette='viridis', hue='country')
plt.title("Countries Contributing Content to Netflix")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Which directors have the most titles on Netflix?

director_counts = data['director'].value_counts()

# print(director_counts)

d = director_counts.head(30).reset_index()
d.columns = ['director', 'count']

plt.figure(figsize=(14, 7))
sns.barplot(x='director', y='count', data=d, palette='magma', hue='director')
plt.title("Directors with Most Titles on Netflix")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
