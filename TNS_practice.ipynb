```python
name = input("Enter your name : ")
age = int(input("Enter your age: "))
nSub = int(input("Enter No. of Subjects you have: "))



StuName = []
StuSubj = set()
StuAge = ()
Sage = []
StuMarks = {}

StuName.append(name)
Sage.append(age)
StuAge = tuple(Sage)

for i in range (nSub):
  SubName = input("Enter Subject Name: ")
  StuSubj.add(SubName)

for namm in StuSubj:
  SubMarks = int(input(f"Enter {namm} subject marks: "))

  StuMarks[namm] = SubMarks


print("----------------STUDENT REPORT CARD----------------------")
print(f"Name of the Student: {name}")
print(f"Age : {age}")
for namm,marks in StuMarks.items():
  print(f"{namm} | {marks}")
total = sum(StuMarks.values())
avg = total / len(StuMarks)

print("Total Marks :", total)
print("Average Marks :", avg)
```
