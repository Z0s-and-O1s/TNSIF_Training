```python
# Task 1
# Artimatic operators
a = 10
b = 20
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
```

```python
# logical
a = True
b = False

print(a and b)
print(a or b)
print(not a)
print(not b)
```

```python
# Comparition
a = 10
b = 20

print("a == b :", a == b)
print("a != b :", a != b)
print("a > b :", a > b)
print("a < b :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
```

```python
# Assignment
a = 10

a += 5
print("a += 5 :", a)

a -= 3
print("a -= 3 :", a)

a *= 2
print("a *= 2 :", a)

a /= 4
print("a /= 4 :", a)

a %= 3
print("a %= 3 :", a)
```
