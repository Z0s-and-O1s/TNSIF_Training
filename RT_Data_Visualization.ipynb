```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Presentation-TASK - Sheet1.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data = data.drop(columns='Marks')
data
```

```python
data = data.dropna()
data
```

```python
data.isnull().sum()
```

```python
data['Total_Marks(25)'] = (data['Technical knowledge(10)']+data['Communication(10)']+data['Attrire(5)'])
data
```

```python
data.to_csv('processed_data.csv', index=False)
```

```python
import plotly.express as px
```

```python
# 1. Top 5 students based on Total Score (Technical + Communication + Attire)

Sorted_marks = data.sort_values(by='Total_Marks(25)', ascending=False).head(5)
display(Sorted_marks)

fig = px.bar(
    Sorted_marks,
    x='Participant Name',
    y='Total_Marks(25)',
    title='Top 5 Students Based on Total Marks',
    color='Total_Marks(25)',
    text='Total_Marks(25)'
)

fig.update_layout(
    xaxis_title="Student Name",
    yaxis_title="Total Marks"
)

fig.show()
```

```python
# 2. Top 5 students in Technical

Sorted_Tech = data.sort_values(by='Technical knowledge(10)', ascending=False).head(5)
display(Sorted_Tech)

fig = px.bar(
    Sorted_Tech,
    x='Participant Name',
    y='Technical knowledge(10)',
    color='Technical knowledge(10)',
    text='Technical knowledge(10)',
    title='Top 5 Students in Technical'
)

fig.update_traces(textposition='outside')

fig.update_layout(
    xaxis_title='Student Name',
    yaxis_title='Technical Marks'
)

fig.show()
```

```python
# 3. Top 5 students in Communication

Sorted_Comm = data.sort_values(by='Communication(10)', ascending=False).head(5)
display(Sorted_Comm)

fig = px.bar(
    Sorted_Comm,
    x='Participant Name',
    y='Communication(10)',
    color='Communication(10)',
    text='Communication(10)',
    title='Top 5 Students in Communication'
)

fig.update_traces(textposition='outside')

fig.update_layout(
    xaxis_title='Student Name',
    yaxis_title='Communication Marks'
)

fig.show()
```

```python
# 4. Top 5 students in Attire

Sorted_Att = data.sort_values(by='Attrire(5)', ascending=False).head(5)
display(Sorted_Att)

fig = px.bar(
    Sorted_Att,
    x='Participant Name',
    y='Attrire(5)',
    color='Attrire(5)',
    text='Attrire(5)',
    title='Top 5 Students in Attire'
)

fig.update_traces(textposition='outside')

fig.update_layout(
    xaxis_title='Student Name',
    yaxis_title='Attire Marks'
)

fig.show()
```

```python
fig = px.box(
    data,
    y=[
        'Technical knowledge(10)',
        'Communication(10)',
        'Attrire(5)'
    ],
    title='Marks Distribution'
)

fig.update_layout(
    template='plotly_white'
)

fig.show()
```

```python
print("="*60)
print("DASHBOARD SUMMARY")
print("="*60)

# Highest Total Marks
highest_total = data['Total_Marks(25)'].max()
top_total = data[data['Total_Marks(25)'] == highest_total]

print(f"\nHighest Total Marks obtained        : {highest_total}")
print("Topper(s):")
for name in top_total['Participant Name']:
    print(f"   • {name}")

# Highest Technical Marks
highest_tech = data['Technical knowledge(10)'].max()
top_tech = data[data['Technical knowledge(10)'] == highest_tech]

print(f"\nHighest Technical Marks obtained    : {highest_tech}")
print("Topper(s):")
for name in top_tech['Participant Name']:
    print(f"   • {name}")

# Highest Communication Marks
highest_comm = data['Communication(10)'].max()
top_comm = data[data['Communication(10)'] == highest_comm]

print(f"\nHighest Communication Marks obtained: {highest_comm}")
print("Topper(s):")
for name in top_comm['Participant Name']:
    print(f"   • {name}")

# Highest Attire Marks
highest_att = data['Attrire(5)'].max()
top_att = data[data['Attrire(5)'] == highest_att]

print(f"\nHighest Attire Marks obtained       : {highest_att}")
print("Topper(s):")
for name in top_att['Participant Name']:
    print(f"   • {name}")
```
