##Artificial Intelligence (AI)

Artificial Intelligence (AI) is a branch of computer science that aims to create systems capable of performing tasks that normally require human intelligence.

- These tasks include:

1. Understanding language
2. Recognizing images and speech
3. Making decisions
4. Solving problems
5. Learning from experience

In simple terms, AI is the science of making machines think and act intelligently.

- Examples of AI
- Siri and Google Assistant answering questions.
- ChatGPT generating human-like responses.
- Self-driving cars making driving decisions.
- Chess-playing programs competing against humans.


##Machine Learning (ML)
------------------------
Machine Learning is a subset of Artificial Intelligence.

Instead of programming every rule manually, machine learning enables computers to learn patterns from data and improve their performance automatically.

####Traditional Programming
----------------------------
Input + Rules → Output

Example:

Marks = 95

Rule:
If marks ≥ 90, assign Grade A.

Output:
Grade A

Here, humans explicitly provide the rules.

####Machine Learning Approach
----------------------------
Input Data + Correct Answers → Learning Algorithm → Model

Once trained, the model can predict outcomes for new data.

Example:

Suppose we provide information about houses:

| Area(sq.ft) |	Price |
-------------------------
| 1000	  | 20 Lakhs |

| 1500	| 30 Lakhs |

| 2000	| 40 Lakhs |

After learning from these examples, the model can predict the price of a new house.

----------------------------

##Deep Learning (DL)

Deep Learning is a subset of Machine Learning that uses Artificial Neural Networks inspired by the structure of the human brain.

Deep learning works exceptionally well when dealing with large amounts of data such as:

- Images
- Videos
- Audio
- Text

It automatically learns features from data without requiring extensive manual feature engineering.

- Examples of Deep Learning

1. Face recognition systems.
2. Speech recognition in Alexa and Siri.
3. Self-driving cars identifying roads and traffic signs.
4. Google Translate.
5. ChatGPT and Large Language Models.

#Types of Machine Learning
1. Supervised Learning
2. Unsupervised Learning
3. Semi-Supervised Learning
4. Reinforcement Learning

#1. Supervised Learning
Definition

Supervised Learning is a type of Machine Learning in which the model learns from labeled data.

Labeled data means the input data is provided along with the correct output.

The model learns the relationship between inputs (features) and outputs (labels) and then predicts outcomes for new, unseen data.

- Example
####Email Spam Detection

Input:

Email Content

Output:

Spam or Not Spam

####Student Performance Prediction

Input:

Attendance
Study Hours
Previous Marks

Output:

Pass/Fail

###Algorithms
- Regression Algorithms - Used when output is continuous.
- Classification Algorithms - Used when output belongs to categories.

#2. Unsupervised Learning
Definition

Unsupervised Learning is a type of Machine Learning where the model learns from data without labeled outputs.

The objective is to discover hidden patterns, relationships, or groups within the data.

Only input data is provided.

- Example

Suppose a shopping mall has customer data but no information about customer categories.

The model automatically groups customers based on purchasing behavior.

###Algorithms
- Clustering Algorithms-Used to group similar data points.
- Dimensionality Reduction Algorithms-Used to reduce the number of features.
- Association Rule Learning-Used to find relationships among items.

#3. Semi-Supervised Learning
Definition

Semi-Supervised Learning combines both labeled and unlabeled data.

A small portion of the dataset is labeled, while a large portion remains unlabeled.

This approach reduces the cost and effort of labeling large datasets.

- Example

Suppose we have:

500 labeled images of cats and dogs
10,000 unlabeled images

The model learns from both datasets to improve prediction accuracy.

#4. Reinforcement Learning
Definition

Reinforcement Learning is a type of Machine Learning where an agent learns by interacting with an environment.

Real-Life Examples

- Self-Driving Cars

The car learns the best driving actions through rewards and penalties.

- Chess and Board Games

AI learns winning strategies.
