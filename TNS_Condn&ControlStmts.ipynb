```python
#to check the number is prime or not
num = int(input("Enter number"))
count=0
for i in range(1,num+1):
  if(num%i==0):
    count+=1
if(count==2):
  print("its a prime number")
else:
  print("not a prime")
```

```python
#program on palindrome
num = 123
temp = num
reverse = 0
while num>0:
  reminder =num%10
  reverse = reverse*10+reminder
  num = num//10
print(reverse)
if(temp==reverse):
  print("palindrome")
else:
  print("Not a palindrome")
```

```python
#Armstrong number
num = 944
sum = 0
temp = num
l = len(str(num))
#print(l)
for i in range(l+1):
  rem = num%10
  sum = sum+(rem**l)
  num = num//10
print(sum)
if(temp==sum):
  print("Armstrong")
else:
  print("Not a armstrong")
```

```python
# Reverse of a string

s = input("Enter a string: ")

reverse = ""

for ch in s:
    reverse = ch + reverse

print("Reversed string:", reverse)
```

```python
# Reverse of a number
n = 1234
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print(rev)
```

```python
# Max of 3 munbers
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Largest =", a)
elif b > c:
    print("Largest =", b)
else:
    print("Largest =", c)
```
